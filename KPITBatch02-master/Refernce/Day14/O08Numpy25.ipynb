```python
import numpy as np

# get the current date and time

date_time = np.datetime64("now")

print(date_time)
```

```python
# to get the current date 

today_date = np.datetime64('today', 'D')
print(today_date)
```

```python
# datetime64 for different time units

year = np.datetime64('2023', "Y")
print(f"Year :{year}")

month = np.datetime64('2023-04', "M")
print(f"Month :{month}")

day = np.datetime64('2023-04-29', "D")
print(f"Day :{day}")

hour = np.datetime64("2023-11-25T08:52:43", "h")
print(f"hour :{hour}")

minute = np.datetime64("2023-11-25T08:52:43", "m")
print(f"minute :{minute}")

seconds = np.datetime64("2023-11-25T08:52:43", "s")
print(f"seconds :{seconds}")
```

```python
# converting date time object datetime64 to python date time

from datetime import datetime

dt64 = np.datetime64("2023-11-25T08:52:43")

# convert to python date time
dt = dt64.astype(datetime)
print(dt)
```

```python
# from python date time to datetime64 object

dt = datetime(2023, 4, 29, 12, 34, 56)
print(dt)

dt64 = np.datetime64(dt)
print(dt64)
```

```python
# create a range of dates - datetime64

dates = np.arange('2023-04-01', '2023-04-11', dtype="datetime64[D]")
print(dates)
```

```python
# arithematic operations on datetime

today = np.datetime64("today")
print(today)

tmrw = today + np.timedelta64(1, "D")
print(f"tomorrow :{tmrw}")

date1 = np.datetime64('2023-05-01')
date2 = np.datetime64("2023-04-01")

num_days = date1 - date2

print(num_days)
```

```python
# scatter graph for data visualization

import matplotlib.pyplot as plt

car = np.array(['BMW', "Audi", 'Benz', 'Honda', 'Toyota', 'Lexus'])
print(f"car :{car}")

weight = np.array([0.5, 1.7, 2, 2, 2.3, 3])
print(f"weight :{weight}")

colors = np.array([6, 7, 8, 9, 10, 11])
sizes = np.array([20, 40, 60, 80, 100, 150])

plt.scatter(car, weight, c=colors, s=sizes)
plt.show()
```
