```python
# interpolation

import numpy as np

day = [2, 4, 7]

petrol_prc =[95, 100, 103]

day3_prc = np.interp(3, day, petrol_prc)
print(f"Petrol price on Day - 3 :{day3_prc}")

day6_prc = np.interp(6, day, petrol_prc)
print(f"Petrol price on Day - 6 :{day6_prc}")
```

```python
day = [2, 4, 7, 10]
petrol_prc = [95, 97, 100, 102]

inter_days = [1, 3, 5, 6, 8, 9]
inter_prc = np.interp(inter_days, day, petrol_prc)

print(f"Interpolated Price :{inter_prc}")
```

```python
import matplotlib.pyplot as plt
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 3, 4, 5, 7, 8])

# generate evenly spaced values between the min and max x values
x_interp = np.linspace(x.min(), x.max(), 100)

y_interp = np.interp(x_interp, x, y)

# plot the original data points and the interpolated values
plt.plot(x, y, 'bo')
plt.plot(x_interp, y_interp, 'r-')

plt.show()
```
